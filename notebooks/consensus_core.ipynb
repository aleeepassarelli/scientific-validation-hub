````python
# consensus_core.py
# M√≥dulo de Valida√ß√£o Cruzada de Inten√ß√£o (Human-in-the-Loop)
# Compara outputs de m√∫ltiplos LLMs para validar a solidez de uma Inten√ß√£o Alg√©brica.

from sentence_transformers import SentenceTransformer, util
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carrega modelo (Reutiliza se j√° estiver na mem√≥ria)
# Usamos o 'all-mpnet-base-v2' pois ele √© o melhor para capturar nuances sem√¢nticas densas
model_consensus = SentenceTransformer('all-mpnet-base-v2')

def run_consensus_engine(intent_anchor: str, model_outputs: dict):
    """
    Analisa o consenso entre diferentes LLMs sobre uma mesma Inten√ß√£o.

    Args:
        intent_anchor (str): A descri√ß√£o l√≥gica/alg√©brica da inten√ß√£o.
        model_outputs (dict): {'GPT-4': 'texto...', 'Claude': 'texto...', 'Llama': 'texto...'}
    """
    print(f"\n{'='*70}")
    print(f"üó≥Ô∏è MOTOR DE CONSENSO DE INTEN√á√ÉO")
    print(f"{'='*70}")

    # 1. Vetoriza√ß√£o
    anchor_emb = model_consensus.encode(intent_anchor, convert_to_tensor=True)
    output_embs = {k: model_consensus.encode(v, convert_to_tensor=True) for k, v in model_outputs.items()}

    stats = []

    # 2. An√°lise de Ader√™ncia (Output vs Inten√ß√£o Original)
    print("\n1. ADER√äNCIA √Ä √ÇNCORA (Inten√ß√£o Original):")
    for name, emb in output_embs.items():
        score = util.cos_sim(anchor_emb, emb)[0][0].item()
        stats.append({'Model': name, 'Adherence': score})
        print(f"   üîπ {name}: {score:.4f}")

    # 3. Matriz de Consenso (Modelo vs Modelo)
    # Se o GPT concorda com o Claude, a l√≥gica √© s√≥lida.
    print("\n2. MATRIZ DE CONSENSO (Coer√™ncia Inter-Modelos):")
    model_names = list(model_outputs.keys())
    matrix_size = len(model_names)
    consensus_matrix = np.zeros((matrix_size, matrix_size))

    for i in range(matrix_size):
        for j in range(matrix_size):
            sim = util.cos_sim(output_embs[model_names[i]], output_embs[model_names[j]])[0][0].item()
            consensus_matrix[i][j] = sim

    # Exibir Matriz textual
    df_matrix = pd.DataFrame(consensus_matrix, index=model_names, columns=model_names)
    print(df_matrix.round(4))

    # 4. C√°lculo do Veredito
    mean_consensus = (np.sum(consensus_matrix) - matrix_size) / (matrix_size * (matrix_size - 1)) # M√©dia fora da diagonal
    min_adherence = min([s['Adherence'] for s in stats])

    print(f"\n{'-'*30}")
    print(f"üìä RESULTADOS FINAIS:")
    print(f"   M√©dia de Consenso: {mean_consensus:.4f} (Qu√£o iguais s√£o os outputs?)")
    print(f"   Ader√™ncia M√≠nima:  {min_adherence:.4f} (O pior modelo entendeu?)")

    # L√≥gica de Aprova√ß√£o (Thresholds calibrados para mpnet)
    # Consenso > 0.85 = As IAs concordam quase totalmente.
    # Ader√™ncia > 0.60 = Seguiram a instru√ß√£o complexa.
    if mean_consensus > 0.85 and min_adherence > 0.60:
        print("\n‚úÖ VEREDITO: VALIDATED (GOLD BADGE)")
        print("   Esta inten√ß√£o √© robusta, inequ√≠voca e port√°til entre modelos.")
        return "PASS"
    elif mean_consensus > 0.75:
        print("\n‚ö†Ô∏è VEREDITO: WARNING (AMBIGUITY DETECTED)")
        print("   Os modelos divergiram na interpreta√ß√£o. Refine a √°lgebra.")
        return "WARN"
    else:
        print("\n‚ùå VEREDITO: FAILED (LOGIC BREAK)")
        print("   A inten√ß√£o gerou alucina√ß√µes distintas. L√≥gica inst√°vel.")
        return "FAIL"

# --- EXEMPLO DE USO (Copie e cole isso numa c√©lula de teste) ---
"""
intent_desc = "Uma fun√ß√£o Python que calcula Fibonacci de forma recursiva, mas com cache (memoization) manual."

outputs_teste = {
    'GPT-4': "def fib(n, cache={}): if n in cache: return cache[n]...",
    'Claude': "Aqui est√° uma implementa√ß√£o recursiva com dicion√°rio de cache...",
    'DeepSeek': "Use um dict para armazenar resultados pr√©vios na recurs√£o..."
}

run_consensus_engine(intent_desc, outputs_teste)
"""
````
